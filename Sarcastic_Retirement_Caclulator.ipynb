# Initialize Environment

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Define Function

```python
def retirement_goal(net_worth):
    """
    Computes the retirement goal based on the given net worth.

    Parameters:
    net_worth (float): The current net worth of the individual.

    Returns:
    float: The retirement goal amount.
    """
    if net_worth < 1_000_000:
        return 1_000_000
    else:
        return 2 * net_worth
```

# Run Analysis

```python
# Make a plot using the function above
net_worths = np.linspace(0, 5000000, 10000)
retirement_goals = np.array([retirement_goal(net_worth) for net_worth in net_worths])
```

# Plot

```python
plt.figure(figsize=(10, 6))
plt.plot(net_worths / 1e6, retirement_goals / 1e6, linestyle='-', color='blue', linewidth=2)
plt.fill_between(net_worths / 1e6, retirement_goals / 1e6, color='blue', alpha=0.8)
plt.title("Net Worth vs. Retirement Goal", fontsize=16)
plt.xlabel("Net Worth ($1M)", fontsize=16)
plt.ylabel("Retirement Goal ($1M)", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.xlim(0, 5)
plt.ylim(0, 10)
plt.show()
```

# Optional: Enter Numerical Value

```python
def main():
    try:
        net_worth = float(input("Enter the person's current net worth: $"))
        goal = retirement_goal(net_worth)
        print(f"The retirement goal for this person is: ${goal:,.2f}")
    except ValueError:
        print("Please enter a valid number for net worth.")
```

```python
if __name__ == "__main__":
    main()
```
